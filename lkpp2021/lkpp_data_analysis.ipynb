## Overview of Data Cleaning
- Columns will be renamed 
- Unnecessary columns will be removed
- Lokasi (location) column values will be split into 3 (province,city,region)
- Special characters will be removed from cells

```python
def rename_columns(df, new_names_list):
    
    """Takes a DataFrame that needs to be renamed and a list of the new
        column names, and returns the renamed DataFrame. The 
        number of columns in the df matches the list length exactly,
        or function will not work as intended."""
    
    rename_dict = dict(zip(df.columns, new_names_list))
    df = df.rename(mapper=rename_dict, axis=1)
    
    return df
```

```python
#making a copy of the original dataframe
translated_df = df.copy()



#creating a list of the new column names and passing it as an array
new_names = ['month_id','total','loc_id','kldi_id', 'method', 'description', 'regional_institute', 'id_ref',
             'location', 'Type', 'time', 'method_id', 'ID', 'type_id', 'work_unit']

new_column_names = np.array(new_names)

translated_df = rename_columns(translated_df, new_column_names)
```

```python
def remove_columns(dataset, cols):
    return dataset.drop(cols, axis=1)
```

```python
columns_drop = ['loc_id',
                'kldi_id',
                'id_ref',
                'time',
                'method_id']

new_df = remove_columns(translated_df, columns_drop)

new_df.head(5)
```

```python
def plot_double(x, height, leftlabel, rightlabel, xlabel=None, ylabel=None, ylocs=None, yticks=None,ylim=None, title = "test"):
    plt.figure(figsize=(7,7))
    plt.bar(x, height)
    plt.text(0-.1,height[0]+(max(height)*0.05), str(round(100*height[0]/sum(height),1))+"%", fontsize=14)
    plt.text(1-.1,height[1]+(max(height)*0.05), str(round((100*height[1]/sum(height)),1))+"%", fontsize=14)

    plt.title(title, fontsize=14)
    plt.xticks([0,1], [leftlabel, rightlabel], fontsize=13)
    if xlabel:
        plt.xlabel(xlabel,fontsize=13)
    if ylabel:
        plt.ylabel(ylabel,fontsize=13)
    if ylocs:
        plt.yticks(ylocs, yticks)
    if ylim:
        plt.ylim(0,ylim)

    plt.show()
```

# <center>Cleaning and Analying Entry Descriptions with NLTK</center>

```python
lkpp.head(1)
```

```python
goods_only = lkpp[lkpp["Type"] == 'Barang']

goods_only['description'] = goods_only['description'].str.lower()
```

```python
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize, RegexpTokenizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
import warnings
warnings.filterwarnings("ignore")


st = StemmerFactory()
stemmer = st.create_stemmer()



def remove_stopwords(desc):
    
    factory = StopWordRemoverFactory()
    sw = set(factory.get_stop_words())

    return[item for item in desc if item not in sw]



# Remove all punctuation from the job description
def remove_punc(desc):
    
    """Removes all punctuation from the set of string."""
    
    punctokenizer = RegexpTokenizer(r'\w{3,}')
    return punctokenizer.tokenize(desc)


def stem_description(desc):
    """This function takes the list of words and reduce inflection in words to their root forms."""
    
    
    st = StemmerFactory()
    stemmer = st.create_stemmer()
    
    stem_result = []
    
    for word in desc:
        stem_result.append(stemmer.stem(word))
        
        
    return stem_result

```

## Most Common Words in Descriptions

```python
descriptions_all = lkpp['description'].apply(remove_punc).apply(remove_stopwords)
```

```python
most_common_words = nltk.FreqDist([word for singledescription in descriptions_all.tolist()
                                   for word in singledescription]).most_common(30)

x_freq = [x[0] for x in most_common_words]
y_freq = [y[1] for y in most_common_words]

plt.figure(figsize=(10, 10))
sns.barplot(y=x_freq, x=y_freq, color='#008FD5')
plt.title('Most Common Words in Descriptions')
plt.show()
```


#### Using Bigrams to search relevant pair of words occurence. Majority of our relevant search words are in pairs i.e
<br>`tempat sampah`, `gerobak sampah`, `truk sampah`, etc.

```python
token_words = goods_only['description'].apply(remove_punc).apply(remove_stopwords)


bg_words = [word for singledescription in token_words.tolist()
              for word in singledescription]


bgs = nltk.bigrams(bg_words)
tokenized_bg = nltk.FreqDist(bgs)
tokenized_bg.most_common(10)
```

```python
tempat_sampah     = tokenized_bg['tempat','sampah']
gerobak_sampah    = tokenized_bg['gerobak','sampah']
motor_sampah      = tokenized_bg['motor', 'sampah']
mobil_sampah      = tokenized_bg['mobil', 'sampah']
roda_tiga         = tokenized_bg['roda','tiga']
truk_sampah       = tokenized_bg['truk','sampah'] + tokenized_bg['truck','sampah']
dump_truk         = tokenized_bg['dump','truk'] + tokenized_bg['dump','truck']
arm_roll          = tokenized_bg['arm','roll']
kontainer_sampah  = tokenized_bg['container','sampah'] + tokenized_bg['kontainer','sampah']
plastik_sampah    = tokenized_bg['plastik','sampah']
pakaian_kerja     = tokenized_bg['pakaian','kerja']
penghancur_sampah = tokenized_bg['penghancur','sampah']
bak_sampah        = tokenized_bg['bak','sampah']


occurences = [tempat_sampah, gerobak_sampah, motor_sampah, mobil_sampah, roda_tiga, truk_sampah, dump_truk, arm_roll, kontainer_sampah, plastik_sampah, pakaian_kerja, penghancur_sampah, bak_sampah]
    
words = ['tempat_sampah','gerobak_sampah','motor_sampah','mobil_sampah','roda_tiga','truk_sampah','dump_truk',
         'arm_roll','kontainer_sampah','plastik_sampah','pakaian_kerja','penghancur_sampah','bak_sampah']


new = {'words':words,'occurences':occurences}
words_df = pd.DataFrame(new)

words_df = words_df.astype({"occurences": int})
```

## Most Common Waste-Management Hardware/Equipments on Demand

```python
plt.figure(figsize=(10, 10))
sns.barplot(y=words_df.words, x=words_df.occurences, 
            color='#008FD5',
            order=words_df.sort_values('occurences', ascending=False).words)
plt.title('Relevant words of interest in Descriptions')
plt.show()
```
